```python
import pandas as pd
import matplotlib.pyplot as plt
weather = pd.read_csv("https://raw.githubusercontent.com/alanjones2/dataviz/master/london2018.csv")
weather
```

```python
weather.plot(x="Month", y="Sun")
```

```python
weather.plot.bar(x="Month", y="Sun")
```

```python
weather.plot(x="Month", y=["Sun", "Rain", "Tmax", "Tmin"])
```

```python
weather.plot.bar(x="Month", y="Rain", color=["black"])
```

```python
weather.plot.bar(x="Month", y=["Sun", "Rain", "Tmax", "Tmin"], color=["#E58874", "#944737","Blue", "Yellow"])
```

```python
weather.plot.scatter(x="Month", y="Sun")
```

```python
weather.plot(x="Month", y=["Sun"], color=["Green"], linestyle=":", linewidth=4, markerfacecolor="black", markersize="15", marker="*")
plt.xlabel("Month of the year 2018 (London CIty)")
plt.ylabel("Value of the rain")
plt.title("Rain value vs Months of the year 2018")
plt.show()
```

```python
weather.plot.bar(x="Month", y=["Sun", "Rain", "Tmax"], color=["Pink", "olive","Blue"], layout=(2,2), subplots=True)
```

```python
weather.plot.bar(x="Month", y=["Sun", "Rain", "Tmax","Tmin"], color=["Purple", "Orange","Blue","Yellow"], layout=(2,2), subplots=True)
```

```python
weather.plot.pie(subplots=True,layout=(2,3) )
```
