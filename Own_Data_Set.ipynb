```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.catplot(x= "video_view_count", y = "video_like_count", data =df, kind = "box", aspect = 2)
plt.title("Boxplot for video_view_count vs video_like_count")
plt.show()
```

```python
value_column='video_like_count'
outliers= df[(df[value_column] < df[value_column].mean() - 3 * df[value_column].std())|
             (df[value_column] > df[value_column].mean() + 3 * df[value_column].std())]
print("Outliers: \n")
outliers
```
