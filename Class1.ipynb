```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.catplot(x= "ocean_proximity", y = "population", data =df, kind = "box", aspect = 2)
plt.title("Boxplot for ocean proximity vs population")
plt.show()
```

```python
value_column='population'
outliers= df[(df[value_column] < df[value_column].mean() - 3 * df[value_column].std())|
             (df[value_column] > df[value_column].mean() + 3 * df[value_column].std())]
print("Outliers: \n")
outliers
```
