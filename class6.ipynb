```python
# Question 26: What is the survival rate based on the cabin class (if available)?

# survival_rate_by_cabin_class = df.groupby(['Cabin', 'Pclass'])['Survived'].mean().reset_index()
# survival_rate_by_cabin_class

df['CabinClass'] = df['Cabin'].str[0]


df['CabinClass'].fillna('Unknown', inplace=True)


df.pivot_table(index='CabinClass', values='Survived', aggfunc='mean')


```
