```python
column = ['alcohol', 'flavanoids', 'malic_acid']

for i in range(0,3,1):
  sns.catplot(x="target", y=column[i], data=df,kind="box",aspect=1.5)
  if column[i]=='alcohol':
    plt.title("boxplot for target vs alcohol")
  elif column[i]=='flavanoids':
    plt.title("boxplot for target vs flavanoids")
  elif column[i]=='malic_acid':
    plt.title("boxplot for target vs malic_acid")

  plt.show()
```

```python
sns.scatterplot(x="proline", y="flavanoids", hue="target",data=df,palette="Dark2", s=80)
plt.title("Relationship between proline, flavanoids and target")
plt.show()
```
